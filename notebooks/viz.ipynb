```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
```

```python
df_energy = pd.read_csv("../data/processed/energy_cleaned.csv")
```

```python
df_geo = pd.read_csv("../data/processed/country_geo.csv")
```

```python
print(df_energy.columns)
```

```python
print(df_geo.columns)
```

```python
df_merged = df_energy.merge(
    df_geo,
    left_on="Entity",
    right_on="Country",
    how="left"
)

df_merged = df_merged.drop(columns=["Entity"])

print(df_merged.columns)
```

```python
print(df_merged.info())
```

```python
print(df_merged.head())
```

```python
df_merged = df_merged.drop(columns=["geometry"])
df_merged["geometry"] = df_merged["geometry_wkt"].apply(wkt.loads)
df_merged = df_merged.drop(columns=["geometry_wkt"])
```

```python
gdf = gpd.GeoDataFrame(df_merged, geometry="geometry")
```
