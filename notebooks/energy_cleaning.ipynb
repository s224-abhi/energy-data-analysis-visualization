```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df_energy = pd.read_csv("../data/raw/global-data-on-sustainable-energy.csv")
print(df_energy.info())
```

```python
df_energy["Density(P/Km2)"] = (
    df_energy["Density(P/Km2)"].str.replace(",", "").astype(float)
)
df_energy['Density(P/Km2)'].dtype
```

```python
df_info = df_energy.describe().transpose()
missing_values = df_energy.isna().sum()
print(missing_values)
```

```python
# Drop columns with a high number of missing values
df_energy.drop(
    columns=[
        "Financial flows to developing countries (US $)",
        "Renewables (% equivalent primary energy)",
        "Renewable-electricity-generating-capacity-per-capita",
    ],
    inplace=True,
)
```

```python
# Fill missing values with mean
columns_to_fill_mean = [
    "Access to clean fuels for cooking",
    "Renewable energy share in the total final energy consumption (%)",
    "Electricity from nuclear (TWh)",
    "Energy intensity level of primary energy (MJ/$2017 PPP GDP)",
    "Value_co2_emissions_kt_by_country",
    "gdp_growth",
    "gdp_per_capita",
]
df_energy[columns_to_fill_mean] = df_energy[columns_to_fill_mean].apply(
    lambda x: x.fillna(x.mean())
)
```

```python
df_energy = df_energy.dropna()
```

```python
print(df_energy.isna().sum())
```

```python
df_energy.describe()
```

```python
df_energy = df_energy[
    [
        "Entity",
        "Year",
        "Access to electricity (% of population)",
        "Access to clean fuels for cooking",
        "Renewable energy share in the total final energy consumption (%)",
        "Electricity from fossil fuels (TWh)",
        "Electricity from nuclear (TWh)",
        "Electricity from renewables (TWh)",
        "Low-carbon electricity (% electricity)",
        "Primary energy consumption per capita (kWh/person)",
        "Density(P/Km2)",
        "gdp_per_capita",
    ]
]
```

```python
df_energy['Total_electricity'] = (
    df_energy['Electricity from fossil fuels (TWh)'] +
    df_energy['Electricity from nuclear (TWh)'] +
    df_energy['Electricity from renewables (TWh)']
)

df_energy['Fossil_ratio'] = df_energy['Electricity from fossil fuels (TWh)'] / df_energy['Total_electricity']
df_energy['Renewable_ratio'] = df_energy['Electricity from renewables (TWh)'] / df_energy['Total_electricity']
df_energy['Nuclear_ratio'] = df_energy['Electricity from nuclear (TWh)'] / df_energy['Total_electricity']
```

```python
df_energy['Energy_access_score'] = (
    df_energy['Access to electricity (% of population)'] * 0.6 +
    df_energy['Access to clean fuels for cooking'] * 0.4
)
```

```python
df_energy["Fossil_per_capita"] = (
    df_energy["Electricity from fossil fuels (TWh)"] / df_energy["Density(P/Km2)"]
)
df_energy["Renewable_per_capita"] = (
    df_energy["Electricity from renewables (TWh)"] / df_energy["Density(P/Km2)"]
)
```

```python
# Ensure data is sorted by country and year
df_energy.sort_values(['Entity', 'Year'], inplace=True)

# Calculate year-on-year percentage change for electricity sources
df_energy['Fossil_growth'] = df_energy.groupby('Entity')['Electricity from fossil fuels (TWh)'].pct_change() * 100
df_energy['Renewable_growth'] = df_energy.groupby('Entity')['Electricity from renewables (TWh)'].pct_change() * 100
df_energy['Total_energy_growth'] = df_energy.groupby('Entity')['Total_electricity'].pct_change() * 100

# Fill NaN values for the first year of each country with 0
df_energy['Fossil_growth'] = df_energy['Fossil_growth'].fillna(0)
df_energy['Renewable_growth'] = df_energy['Renewable_growth'].fillna(0)
df_energy['Total_energy_growth'] = df_energy['Total_energy_growth'].fillna(0)

# Optional: view first few rows to verify
df_energy[['Entity', 'Year', 'Fossil_growth', 'Renewable_growth', 'Total_energy_growth']].head(10)
```

```python
df_energy.info()
```

```python
scaler = MinMaxScaler()
df_energy[
    ["Fossil_ratio", "Renewable_ratio", "Nuclear_ratio", "Energy_access_score"]
] = scaler.fit_transform(
    df_energy[
        ["Fossil_ratio", "Renewable_ratio", "Nuclear_ratio", "Energy_access_score"]
    ]
)
```

```python
cols_to_keep = [
    'Entity', 'Year', 'Total_electricity', 'Fossil_ratio', 'Renewable_ratio',
    'Nuclear_ratio', 'Energy_access_score', 'Fossil_per_capita',
    'Renewable_per_capita', 'Fossil_growth', 'Renewable_growth', 'Total_energy_growth'
]

df_energy_plot = df_energy[cols_to_keep].copy()

print(df_energy_plot.describe())
```
